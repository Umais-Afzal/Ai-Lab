```python
def luhn_check(number):
    number = str(number)
    total = 0
    reverse_digits = number[::-1]

    for i in range(len(reverse_digits)):
        digit = int(reverse_digits[i])

        if i % 2 == 1:
            digit = digit * 2
            if digit > 9:
                digit = digit - 9

        total += digit

    return total % 10 == 0


card_number = input("Enter card number: ")

if luhn_check(card_number):
    print(f"{card_number} is a Valid Card Number")
else:
    print(f"{card_number} is an Invalid Card Number")
```

```python
def puntuation_remover(user_input):
    puntuations=',!@#$%^&*()[]~|.''<>"'
    results =" " 
    for char in user_input:
        if char not in puntuations:
            results += char
    return results

_input_ = input("Enter the string")

puntuation_remover(_input_)
```

```python
def text_sort(words):
    words=words.split()
    sorted_words=[]

    while words:
        small_words=min(words)
        sorted_words.append(small_words)
        words.remove(small_words)

    return " ".join(sorted_words)

sent_= "CS AI DS IT SE  "

text_sort(sent_)
```
